```python
# Bai 1
D = {
    "one": 1,
    "two": 2,
    "four": 4,
    "three": 3,
    "five": 5
}
def get_key(val, dict):
    for key, value in dict.items():
        if val == value:
            return key
    return "key doesn't exist"

sorted(D.values())
print([D[get_key(i, D)] for i in sorted(D.values())])
D = {k: v for k, v in sorted(D.items(), key=lambda item: item[1])}
```

```python
# Bai 2
D = {
    "one": 1,
    "two": 2,
    "four": 4,
    "three": 4,
    "five": 5
}

check_dup = {}
for key, value in D.items():
    if value in check_dup:
        check_dup[value].append(key)
    else:
        check_dup[value] = [key]


D = {k: v for k, v in D.items() if len(check_dup[v]) == 1}
D
```

```python
# Bai 5
prices = {
    "banana": 4,
    "apple": 2,
    "orange": 1.5,
    "pear": 3
}

stock = {
    "banana": 6,
    "orange": 32,
    "pear"  : 15,
}

total = {i:prices[i] * stock[i] for i in prices if i in stock}

total.update({i:0 for i in prices if i not in stock})

sorted_total = sorted(total.values())

print(sorted_total)

def get_key(val, dict):
    for key, value in dict.items():
        if val == value:
            return key
    return "key doesn't exist"

for i in sorted_total[::-1]:
    print(get_key(i), round(i), sep="\t")
```

```python
# Bai 7
dictA = { 
    "a": 1,
    "b": 2,
    "c": 3,
    "f": 5,   
}
dictB = {
    "a": 3,
    "d": 2,
    "e": 3,
    "f": 1
}

def create_dict(dictA, dictB):
    dictC = dictA.copy()
    dictC.update(dictB)
    for i in dictA.keys():
        if i in dictB.keys():
            dictC[i] = max(dictA[i], dictB[i])
    return dictC

print(create_dict(dictA, dictB))
```
